```python
import torch
import torch.distributions
from torch.distributions import Normal, MultivariateNormal
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import scale
import torch.nn.functional as F

import numpy as np
import matplotlib.pyplot as plt

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

if torch.cuda.is_available():
    FloatTensor = torch.cuda.FloatTensor
elif not torch.cuda.is_available():
    FloatTensor = torch.FloatTensor
```

```python
def generate_weightuncertainty_data():

    x = np.linspace(-0.35, 0.55, 200)
    x_noise = np.random.normal(0., 0.01, size=x.shape)
    y_noise = np.random.normal(0., 0.1, size=x.shape)

    y = x + 0.3 * np.sin(2 * np.pi * (x + x_noise)) + 0.3 * np.sin(4 * np.pi * (x + x_noise)) + y_noise

    x, y = x.reshape(-1, 1), y.reshape(-1, 1)

    mu = np.array([[-0.3,-0.3],[-0.18, -0.8],[0,0], [0.15, 0.8], [0.35,0.3], [0.55, 0.6]])

    plt.scatter(x, y, alpha=0.5)
    plt.scatter(mu[:,0], mu[:,1], color='red', marker='+', s=100)

    return x, y

X, y = generate_weightuncertainty_data()
X, y = FloatTensor(scale(X)), FloatTensor(scale(y))
```

```python
class GP_InducingPoints(torch.nn.Module):

    def __init__(self, x=None, y=None,
                 num_inducing_points = 6,
                 dim=1):
        super().__init__()

        self.x = x
        self.y = y
        self.dim = dim
        self.num_inducing_points = num_inducing_points

        # init lenght scale and noise estimate
        self.length_scale = torch.nn.Parameter(
            torch.scalar_tensor(0.1))
        self.noise = torch.nn.Parameter(
            torch.scalar_tensor(0.5))

        # init inducing points 
        inducingX = torch.linspace(X.min().item(), X.max().item(), 
                                    self.num_inducing_points).reshape(-1,1) 
        self.inducingX_mu = torch.nn.Parameter(
            inducingX + torch.randn_like(inducingX).clamp(-0.1,0.1))
        self.inducing_y_mu = torch.nn.Parameter(
            FloatTensor(num_inducing_points, dim).uniform_(-0.5,0.5))

        
        
    def KM(self, x1, x2):
        '''compute kernel matrix between x1 and x2'''
        dist = (x1 - x2.T)**2
        scale_dist = dist / (self.length_scale + 0.0001)
        K = torch.exp(-0.5* scale_dist)
        return K
    
    
    def predict(self, x):
        
        # get kernel matrices
        self.KXsX = self.KM(x, self.inducingX_mu)
        self.KXX = self.KM(self.inducingX_mu, self.inducingX_mu)
        self.KXsXs = self.KM(x, x)

        self.KXX_inv = torch.inverse(
            self.KXX + 1e-10*torch.eye(self.KXX.shape[0]))

        # compute mean and covariance for forward prediction
        mu = self.KXsX @ self.KXX_inv @ self.inducing_y_mu
        sigma = (self.KXsXs 
                 - self.KXsX @ self.KXX_inv @ self.KXsX.T 
                 + self.noise*torch.eye(self.KXsXs.shape[0]))

        # output MAP estimate and variance
        return mu, torch.diag(sigma)[:, None]
    
    
    def NMLL(self, X, y):

        # constraint parameters
        self.length_scale.data = self.length_scale.data.clamp( 0.00001, 3.0)
        self.noise.data = self.noise.data.clamp(0.000001,3)

        # compute kernel matrices
        KXsXs = self.KM(X, X)
        KXsX = self.KM(X, self.inducingX_mu)
        KXX = self.KM(self.inducingX_mu, self.inducingX_mu)
        KXX_inv = torch.inverse(KXX + 1e-10*torch.eye(KXX.shape[0]))

        # compute mean and covariance
        mu = KXsX @ KXX_inv @ self.inducing_y_mu
        Sigma = (KXsXs 
                 - KXsX @ KXX_inv @ KXsX.T
                 + self.noise**2*torch.eye(KXsXs.shape[0]))

        p_y = MultivariateNormal(mu.squeeze(), covariance_matrix=Sigma)
        # evaluate the probability of the target values 
        #in the training data set under the distribution of the GP
        mll = p_y.log_prob(y.squeeze()) 
        # add a regularization term to regularize variance
        mll -= 1/( 2 * self.noise**2) * torch.trace(Sigma)

        return -mll
    
    
    def plot(self, _title=""):

        x = torch.linspace(self.x.min()*1.5, self.x.max()*1.5, 200).reshape(-1,1)

        with torch.no_grad():
            mu, sigma = self.forward(x)

        x = x.numpy().squeeze()
        mu = mu.numpy().squeeze()
        sigma = sigma.numpy().squeeze()

        plt.title(_title)
        plt.scatter(self.inducingX_mu.detach().numpy(), self.inducing_y_mu.detach().numpy())
        plt.scatter(self.x.detach().numpy(), self.y.detach().numpy(), alpha=0.1, c='r')
        plt.fill_between(x, mu-3*sigma, mu+3*sigma, alpha = 0.1, color='blue')
        plt.plot(x, mu)
        plt.xlim(self.x.min()*1.5, self.x.max()*1.5)
        plt.ylim(-3,3)
        plt.show()
```

```python
# initialize the GP and plot initial prediction
gp = GP_InducingPoints(X, y)


# use two different learning rates since inducing points
# need to potentially cover a far larger distance than kernel parameters
optim = torch.optim.Adam([{"params": [gp.length_scale, gp.noise], "lr": 0.01},
                          {"params": [gp.inducingX_mu, gp.inducing_y_mu,], "lr":0.1}])

train_loader = DataLoader(TensorDataset(FloatTensor(X), FloatTensor(y)),
                        batch_size=200,
                        shuffle=True,
                        num_workers=1)

gp.plot(_title="Init")

for epoch in range(300):
    for i, (data, label) in enumerate(train_loader):
        optim.zero_grad()

        mll = gp.NMLL(data, label)

        mll.backward()
        optim.step()

        if epoch%(300//10)==0:
            print(f'Epoch: {epoch} \t NMLL:{mll:.2f}'
                  f'\t LS {gp.length_scale:.2f} \t Noise: {gp.noise:.2f}')
            gp.plot(_title=f"Training Epoch {epoch:.0f}")

gp.plot(_title="Post Training")
```
